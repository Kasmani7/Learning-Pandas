```python
student = pd.DataFrame({'Name':['Ahsan','Umer','Ali','Taha','Bilal'],
                       'age': np.random.randint(13,25 , size = [5]),
                       'percentage':np.random.randint(50,80 , size = [5]) })


np.random.seed(5)
```

```python
student
```

```python
def abc(age):
    if age>18:
        return 'Adult'
    else:
        return "Non-Adult"
```

```python
student['Status'] = student['age'].apply(abc) #map
```

```python
student
```

```python
def per(percentage):
    if percentage>65:
        return 'Pass'
    else:
        return "Fail"
```

```python
student['Result'] = student['percentage'].apply(per) #map
```

```python
student
```

```python
student[student.Status == 'Adult']
```

```python
student[(student.Name.str.startswith('A'))]
```

```python
student[(student.Name.str.endswith('l'))]
```
